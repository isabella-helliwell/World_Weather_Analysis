```python
#6.1.4
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# add the code that generates the latitudes and longitudes, 
#but first, they need to be stored so that we can access them later. 
```

```python
# Create a set of random latitude and longitude combinations.
#we'll pack the latitudes (lats) and longitudes (lngs) as pairs by zipping them (lat_lngs) with the zip() function.
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)
lat_lngs
```

```python
#The zip object packs each pair of lats and lngs having the same index in their respective array into a tuple. 
#If there are 1,500 latitudes and longitudes, there will be 1,500 tuples of paired latitudes and longitudes, 
# where each latitude and longitude in a tuple can be accessed by the index of 0 and 1, respectively.
#You can only unzip a zipped tuple once before it is removed from the computer's memory.
```

```python
# unpack the lat_lngs zip object into a list
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs)
```

```python
coordinates
```

```python
#6.1.5 Generate Random world cities
# With the list of random latitudes and longitudes, 
# we'll use the coordinates in our lat_lngs tuple to find the nearest city using Python's citipy module.

```

```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy
```

```python
# Next, use the five pairs of latitudes and longitudes we used from our zip practice to get a city and 
# country code from the citipy module.
```

```python
# Create a list for holding the cities.
cities = []
# Identify the nearest city for each latitude and longitude combination.
for coordinate in coordinates:
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    # If the city is unique, then we will add it to the cities list.
    if city not in cities:
        cities.append(city)
# Print the city count to confirm sufficient count.
```

```python
#6.2.6 Get the City Weather Data

#1. Import our dependencies and initialize counters and an empty list that will hold the weather data.
#2. Loop through the cities list.
#3. Group the cities in sets of 50 to log the process as we find the weather data for each city.
#4. Two counters will be needed here: one to log the city count from 1 to 50, and another for the sets.
#5. Build the city_url or endpoint for each city.
#6. Log the URL and the record and set numbers.
#7. Make an API request for each city.
#8. Parse the JSON weather data for the following:
#9. City, country, and date
#10. Latitude and longitude
#11. Maximum temperature
#-Humidity
#-Cloudiness
#-Wind speed
#12. Add the data to a list in a dictionary format and then convert the list to a DataFram
```

```python
# Import the datetime module from the datetime library.
from datetime import datetime
# Import the requests library.
import requests
# Import the API key.
from config import weather_api_key
```

```python
# Starting URL for Weather Map API Call.
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
print(url)
```

```python
# Loop through all the cities in our list.
#for i in range(len(cities)):

    # Group cities in sets of 50 for logging purposes.
    #if (i % 50 == 0 and i >= 50):
      #  set_count += 1
       # record_count = 1
    # Create endpoint URL with each city.
   # city_url = url + "&q=" + cities[i]
    
 ####----the above code will add the index to the city_url, but we want the actual cities added---##### 
### instaed iterate through the list of cities and retrieve both the index, and the city from the list with enumerate ()###
### the syntax is "for i, item in enumerate(list)"
```

```python
# Loop through all the cities in the list.
city_data = []
# Print the beginning of the logging.
print("Beginning Data Retrieval     ")
print("-----------------------------")
record_count = 1
set_count = 1
for i, city in enumerate(cities):

    # Group cities in sets of 50 for logging purposes.
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 1
    # Create endpoint URL with each city.
    city_url = url + "&q=" + city.replace(" ","+")

    # Log the URL, record, and set numbers and the city.
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    # Add 1 to the record count.
    record_count += 1
# Run an API request for each of the cities.
    try:
        # Parse the JSON and retrieve data.
        city_weather = requests.get(city_url).json()
        # Parse out the needed data.
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        # Convert the date to ISO standard.
        city_date = datetime.utcfromtimestamp(city_weather["dt"]).strftime('%Y-%m-%d %H:%M:%S')
        # Append the city information into city_data list.
        city_data.append({"City": city.title(),
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Country": city_country,
                          "Date": city_date})

# If an error is experienced, skip the city.
    except:
        print("City not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("-----------------------------")
print("Data Retrieval Complete      ")
print("-----------------------------")
```
